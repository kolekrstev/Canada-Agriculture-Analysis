```python
# get the exterior coordinates on the polygon OR Multiplygon (as a list)
car['geomlist']=""
for x, i in enumerate(car["geometry"]):
    #check of its Multiplygon (These are lists of polygons)
    if car["geometry"][x].type=="MultiPolygon":
        list_ext_coord = []
        for k in range(0,len(car["geometry"][x])):
            list_ext_coord.append(list(car["geometry"][x][k].exterior.coords))
        car.at[x,'geomlist'] = list_ext_coord
    # For Polygons only
    else:
        car.at[x,'geomlist'] = list(car["geometry"][x].exterior.coords)
car
```

## Mark the Weather stations to ALL CANADA regions

```python
#### Mark the Weather stations to ALL canada regions
#### Weather stations dataset contains all the weather stations in canada but not grouped by Province
#### so we'll have to loop though all of them to mark the region for each
```

```python
# Check if a coordinate is in a polygon

def point_inside_polygon(lat,lng,poly):
    n = len(poly)
    inside =False
    
    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if lat > min(p1y,p2y):
            if lat <= max(p1y,p2y):
                if lng <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (lat-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or lng <= xinters:
                        inside = not inside
        p1x,p1y = p2x,p2y

    return inside
```

```python
## Loop through all Canada regions and than mark the region in the weather station with the region it belongs
from tqdm import tqdm
## NOTE: Will run for an HOUR
## code needs to be optimized
canada_ws_weatherstats["Region"]=""
for x, i in tqdm(enumerate(canada_ws_weatherstats["Station Name"])):
    for k in car.index:
        if car["geometry"][k].type=="MultiPolygon":
            for y in range(0,len(car["geometry"][k])):
                if point_inside_polygon(canada_ws_weatherstats["Latitude"][x],canada_ws_weatherstats["Longitude"][x],car["geomlist"][k][y]):
                    canada_ws_weatherstats.at[x,"Region"]=car["PROVINCE"][k]+car.at[k,"CARUID"][-2:]
                    continue
        # For Polygons only
        else:
            if point_inside_polygon(canada_ws_weatherstats["Latitude"][x],canada_ws_weatherstats["Longitude"][x],car["geomlist"][k]):
                        canada_ws_weatherstats.at[x,"Region"]=car["PROVINCE"][k]+car.at[k,"CARUID"][-2:]
```

```python
## convert canada_ws_weatherstats to Geopandas Dataframe and 
## then plot both Canada Argiculture regions (car) and canada_ws_weatherstats in the same plot

from shapely.geometry import Point, Polygon
# for Geodataframe we need a geometry feature (in this case it will be a point for the long, lat)
geometry = [Point(xy) for xy in zip(canada_ws_weatherstats["Longitude"], canada_ws_weatherstats["Latitude"])]
geometry[:3]
```

```python
crs={"init":"epsg:4326"}
canada_ws_FINAL = gpd.GeoDataFrame(canada_ws_weatherstats,crs=crs, geometry=geometry)
canada_ws_FINAL.head()
```

```python
# Plot BC Regions 1 and 8 weather stations
fig, ax = plt.subplots(figsize=(20,15))
car.plot(column='CARENAME', ax=ax, legend=True)
canada_ws_FINAL[canada_ws_FINAL["Region"]=="BC01"].plot(ax=ax, markersize=20,color="lime", marker="^", label="BC Region 1")
canada_ws_FINAL[canada_ws_FINAL["Region"]=="BC08"].plot(ax=ax, markersize=20,color="lime", marker="^", label="BC Region 8")
#plt.savefig("Weatherstations_BC_Regions_1_8.pdf", format="pdf")
plt.legend(prop={'size':15})
```

```python
# Plot ON Regions weather stations
fig, ax = plt.subplots(figsize=(20,15))
car.plot(column='CARENAME', ax=ax, legend=True)
canada_ws_FINAL[canada_ws_FINAL["Region"].str.startswith("ON")].plot(ax=ax, markersize=20,color="lime", marker="^", label="ON Regions")
#plt.savefig("Weatherstations_ON_Regions_1-5.pdf", format="pdf")
plt.legend(prop={'size':15})
```
