```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
import mysql.connector
import geopandas as gpd
```

```python
connection = mysql.connector.connect(user='root', password='your_password', host='127.0.0.1', database='field_crops')
cursor = connection.cursor()

add_car = ("INSERT INTO growing_region "
           "(CARUID, province, name, region, boundary) "
           "VALUES (%s, %s, %s, %s, %s)")

for i in range(0, len(car)):

    record = car.iloc[i]

    CARUID = int(record.CARUID)

    province = int(record.PRUID)
    if(province == 10):
        province = 'NL'
    elif(province == 11):
        province = 'PE'
    elif(province == 12):
        province = 'NS'
    elif(province == 13):
        province = 'NB'
    elif(province == 24):
        province = 'QC'
    elif(province == 35):
        province = 'ON'
    elif(province == 46):
        province = 'MB'
    elif(province == 47):
        province = 'SK'
    elif(province == 48):
        province = 'AB'
    elif(province == 59):
        province = 'BC'
    elif(province == 60):
        province = 'YT'
    elif(province == 61):
        province = 'NT'
    elif(province == 62):
        province = 'NU'
    else:
        province = '?'
    
    if(province == '?'):
        print(record.PRUID)
    
    name = record.CARENAME

    region = int(record.CARUID[-2:])
    
    boundary = record.geometry.to_wkt()

    data_car = (CARUID, province, name, region, boundary)
    
    

    cursor.execute(add_car, data_car)

    connection.commit()



cursor.close()
connection.close()

```
