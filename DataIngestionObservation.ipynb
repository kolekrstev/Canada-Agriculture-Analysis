```python
import requests
import time
```

```python
df_clid = df[['CLIMATE_IDENTIFIER', 'LATITUDE', 'LONGITUDE']].copy()
df_clid = df_clid.groupby(['CLIMATE_IDENTIFIER']).max()
df_clid.reset_index(inplace = True)
print(df_clid)
```

```python
def get_radiation_data(latitude, longitude):
    http_get = f'https://power.larc.nasa.gov/api/temporal/monthly/point?parameters=ALLSKY_SFC_PAR_TOT&community=AG&longitude={longitude}&latitude={latitude}&format=JSON&start=2000&end=2020'
    response = requests.get(http_get)
    return(response.json())
```

```python
df_radiation = pd.DataFrame()

print('START: ' + str(datetime.now()))

for i in range(0, len(df_clid)):
    response = get_radiation_data(df_clid.iloc[i].LATITUDE, df_clid.iloc[i].LONGITUDE)
    time.sleep(2) #API limit is 60 requests per minute
    station = response['properties']['parameter']['ALLSKY_SFC_PAR_TOT']
    df_station = pd.DataFrame.from_dict(station, orient = 'index')
    df_station['climate_id'] = df_clid.iloc[i].CLIMATE_IDENTIFIER
    df_radiation = pd.concat([df_radiation, df_station], axis = 0)
    
print('FINISH: ' + str(datetime.now()))
```

```python
df_radiation.reset_index(inplace = True)
df_radiation.rename(columns = { 0 : 'radiation', 'index' : 'date' }, inplace = True)
df_radiation['month'] = df_radiation['date'].str[4:]
df_radiation['year'] = df_radiation['date'].str[:4]
df_radiation.drop(columns = ['date'], inplace = True)
df_radiation
```

```python
print('START: ' + str(datetime.now()))

df_final = pd.DataFrame()
df_subset = pd.DataFrame()
df_radiation_subset = pd.DataFrame()

for k in range(0, len(df_clid)):
    df_subset = df[df['CLIMATE_IDENTIFIER'] == df_clid.iloc[k].CLIMATE_IDENTIFIER].copy()
    df_subset.reset_index(inplace = True, drop = True)
    df_radiation_subset = df_radiation[df_radiation['climate_id'] == df_clid.iloc[k].CLIMATE_IDENTIFIER].copy()
    df_radiation_subset.reset_index(inplace = True, drop = True)
    df_radiation_subset = df_radiation_subset.astype({ 'month' : 'int32' , 'year' : 'int32' })
    for i in range (0, len(df_subset)):
        record = df_radiation_subset[(df_radiation_subset['year'] == int(df_subset.iloc[i].LOCAL_YEAR)) & \
                                     (df_radiation_subset['month'] == int(df_subset.iloc[i].LOCAL_MONTH))]
        #print(record.radiation)
        if(record.empty):
            continue
        else:
            df_subset.loc[i, ['RADIATION']] = float(record.radiation)
            
    df_final = pd.concat([df_final, df_subset], axis = 0)

print('FINISH: ' + str(datetime.now()))
```
